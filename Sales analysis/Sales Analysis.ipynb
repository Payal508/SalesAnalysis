#### Read in updated dataframe

```python
all_data = pd.read_csv('all_data.csv')
all_data.head()
```

### Clean up the data!

#### Drop rows of NaN

```python
nan_df = all_data[all_data.isna().any(axis = 1)]
nan_df

all_data = all_data.dropna(how = 'all')
all_data
```

#### Find 'Or' and delete it

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
```

##### Convert columns to the correct types

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

### Augment data with additional columns 

### Task 2: Add a month column

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

### Task 3: Add a sales column

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

#### Question 1. What was the best month for sales? How much was earned that month?

```python
results = all_data.groupby('Month').sum()
results
```

```python
months = range(1, 13)

plt.bar(months, results['Sales'])
plt.xticks(months)
plt.xlabel('Months')
plt.ylabel('Sales in USD ($)')
plt.show()
```

#### Question 2: What city had the highest number of sales?


### Augment the data with a city column

```python
def get_city(address):
    return address.split(',')[1]

# all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x))

# all_data['City'] = all_data['Purchase Address'].apply(lambda x: x.split(',')[1])

def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x) + ' ' + get_state(x))

all_data.head()
```

```python
results = all_data.groupby('City').sum()
results
```

```python
cities = [city for city, df in all_data.groupby('City')]

plt.bar(cities, results['Sales'])
plt.xticks(cities, rotation = 'vertical', size=8)
plt.xlabel('cities')
plt.ylabel('Sales')
plt.show()
```

#### Question 3: What time should we display advertisements to maximize the likelihood of costumer's buying product?

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data.head()
```

```python
hours = [hour for hour, df in all_data.groupby('Hour')]

plt.plot(hours, all_data.groupby(['Hour']).count())
plt.xticks(hours)
plt.grid()
plt.xlabel('Hour')
plt.ylabel('Number of orders')
plt.show()
```

#### Question 4: What products are most often sold together?

```python
all_data.head()
```

```python
data = all_data[all_data['Order ID'].duplicated(keep = False)]
data['Products'] = data.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
data = data[['Order ID', 'Products']].drop_duplicates()
data.head()
```

```python
count = Counter()

for row in data['Products']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
for key, value in count.most_common(10):
    print(key, value)
```
